## 3. Correlations

```python
df.corr()
```

```python
print(df.columns)
```

```python
# Get all names 
for col_name in df.columns: 
    print(col_name)
```

```python
# Create a subset with only quantitative columns

df_sub = df[['Popularity','Days_since_release','Released_after_2017',
       'USA', 'Explicit_false', 'Explicit_true', 'single', 'Mexico',
       'Australia', 'rock','hip hop','country','k-pop','pop','rap','Top10_dummy']]
```

```python
df_sub.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(df_sub.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(16, 16)) # figure size 
plt.matshow(df_sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_sub.shape[1]), df_sub.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(df_sub.shape[1]), df_sub.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

From this, I am going to omit the countries as it doesnt seem to have much of an impact as much as the genre. 

```python
# Create a subset excluding the country columns

sub = df[['Popularity','Days_since_release','Released_after_2017','Explicit_false','Explicit_true', 'single','rock','hip hop','country','k-pop','pop','rap','Top10_dummy']]
```

```python
# Create a subplot with matplotlib. 

f,ax = plt.subplots(figsize=(10,10))
```

```python
# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.

corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 

# The `annot` argument allows the plot to place the correlation coefficients onto the heatmap.
```

```python
corr.figure
```

The only major correlation I see is that hip hop is correlated with Explicit language even more than rap, going to have to look into the categories that the music is broken out in within the data. 

```python
# Create a subset with types of song messages columns

sub_2 = df[['Popularity','Popu_max','Days_since_release','Released_after_2017','anger','anticipation', 'disgust','fear','joy','sadness','trust','negative','positive','n_words','Top10_dummy','Top50_dummy']]
```

## 6. Categorical Plots

```python
# Use a histogram to visualize the distribution of the variables. 

sns.distplot(df['Popularity'], bins = 20)
```

```python
df.loc[df['Popularity'] < 5000, 'Popularity category'] = 'Low popularity'
```

```python
df.loc[(df['Popularity'] >= 5000) & (df['Popularity'] < 35000), 'Popularity category'] = 'Middle popularity'
```

```python
df.loc[df['Popularity'] >= 35000, 'Popularity category'] = 'High popularity'
```

```python
df['Popularity category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the popularity categories created above

sns.set(style="ticks")
g = sns.catplot(x="Days_since_release", y="Top50_dummy", hue="Popularity category", data=df)
```
